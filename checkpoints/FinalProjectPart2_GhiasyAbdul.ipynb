```python
# Batter Data over the past couple years
from pybaseball import batting_stats
data = batting_stats(2014, 2018)
# data.head()
```

```python
# BA added to dataframe
data['BA'] = (data['H'] / data['AB'])
# data.head()
```

```python
# great batters
highBatter = np.where((data['Age'] < 27) & (data['AB'] > 500) & (data['G'] > 150) & (data['H'] > 150) & (data['HR'] >30))
data.iloc[highBatter[0], :].head()
```

```python
# below average batters
lowBatter = np.where((data['Age'] < 27) & (data['AB'] > 500) & (data['G'] > 150) & (data['BA'] < 0.25) & (data['HR'] < 20))
data.iloc[lowBatter[0], :].head()
```
